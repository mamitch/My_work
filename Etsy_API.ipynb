```python
import requests
import config
import pandas as pd

key=config.etsy_api_key
```

```python
headers = {'X-RateLimit-Limit': '10000'}

response=requests.get(f"https://openapi.etsy.com/v2/listings/active?limit=50&offset=0&api_key={key}", headers=headers)
```

```python
categories=requests.get(f"https://openapi.etsy.com/v2/taxonomy/categories?api_key={key}", headers=headers)

#/categories/:tag
```
